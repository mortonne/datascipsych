### Create a module (1 point)

Create a file in `src/testproject` called `recognition.py`, with the following contents:

```python
"""Functions to analyze recognition memory data."""
import numpy as np


def hit_rate(trial_type, response):
    """
    Calculate hit rate based on trial type and response.

    Args:
        trial_type: An array indicating whether each trial was a "target" 
          or a "lure".
        response: An array indicating whether the response was old (1)
          or new (0).
    
    Returns:
        Hit rate (the fraction of targets with "old" responses).
    """
    hr = np.mean(response[trial_type == "target"])
    return hr
```
