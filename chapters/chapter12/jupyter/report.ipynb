# Introduction

In a simulated recognition memory study, eight participants completed a test of whether they could correctly discriminate between target items (which had been previously studied) and lure items (which had not been studied). We want to calculate common measures of recognition memory performance, the hit rate and false-alarm rate, and examine how they vary across participants.

## Results

First, we import the packages that we will need for data analysis.

```python
import numpy as np
# This will throw an error if the recognition module does not exist yet
from testproject import recognition
```

Next, we load the data file for each participant and calculate the hit rate. We store each participant's hit rate in a NumPy array, which we have initialized to have one element for each participant.

```python
# Initialize a variable to hold hit rate for each subject
n_subj = 8
hit_rate = np.zeros(n_subj)

for i in range(n_subj):
    # Load data for this subject
    subject_number = i + 1
    data_file = f"data/sub-{subject_number:02d}_beh.npz"
    data = np.load(data_file, allow_pickle=True)

    # Calculate hit rate for this subject's data
    # This will throw an error if the hit_rate function does not exist yet
    hr = recognition.hit_rate(data["trial_type"], data["response"])
    hit_rate[i] = hr
```

Finally, we calculate the mean and standard deviation of the hit rate across participants.

```python
m_hr = np.mean(hit_rate)
s_hr = np.std(hit_rate, ddof=1)
print(f"Hit rate: mean {m_hr:.2f}, std. dev. {s_hr:.2f}")
```
